```python
import numpy as np
```

```python
# Create a NumPy array with the probabilities of each outcome
probabilities = np.array([0.5, 0.5])
```

```python
# Generate a random number between 0 and 1
random_number = np.random.rand()
```

```python
# Find the probability that the coin landed heads
probability_heads = probabilities[random_number >= 0.5]
```

```python
# Print the probability
print(probability_heads)
```

```python
toss = np.zeros(100)
```

```python
for i in range(100):
    toss[i]=np.random.randint(0,2)
    
print('Head:', toss.mean())
print('Tail:', 1 - toss.mean())
```

#### Binomial Distribution

```python
# Calculate the probability of getting 7 heads in 10 flips of a coin.
p = 0.5 # Probability of heads
n = 10 # Number of experiments
```

```python
probability = np.random.binomial(n, p)
```

```python
print(probability)
```

#### Poisson Distribution

```python
# Calculate the probability of getting 5
# customers in a store in one hour, given that the
# average number of customers is 5 per hour.
lamb = 20 # Mean number of customers in one hour
```

```python
probability = np.random.poisson(lamb)
print(probability)
```

```python
probs = []
for i in range(10):
    probability = np.random.poisson(lamb)
    probs.append(probability)
    
print(np.mean(probs))
```

```python
sns.kdeplot(probs)
```
