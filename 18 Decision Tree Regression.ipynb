#### Import the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Read the data

```python
df = pd.read_csv('datasets/petrol_consumption.csv')
```

```python
df
```

```python
sns.kdeplot(df['Petrol_Consumption']);
```

```python
df.corrwith(df['Petrol_Consumption'])
```

#### Separate the input and output data

```python
x = df.drop('Petrol_Consumption', axis = 1)
y = df['Petrol_Consumption']
```

```python
x.shape
```

```python
x
```

```python
plt.figure(figsize=(16,9))
plt.subplot(2,2,1)
sns.kdeplot(df['Petrol_tax'])
plt.subplot(2,2,2)
sns.kdeplot(df['Paved_Highways'])
plt.subplot(2,2,3)
sns.kdeplot(df['Average_income'])
plt.subplot(2,2,4)
sns.kdeplot(df['Population_Driver_licence(%)'])
```

#### Cross-validation

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, random_state=0, test_size=0.25)
```

```python
x_train.shape
```

```python
x_test.shape
```

#### Build the model

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor(random_state=0)
```

```python
dt.fit(x_train, y_train)
```

#### Plot the tree

```python
x.columns
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(20,20))
plot_tree(dt, fontsize=7, filled=True, rounded=True,
         feature_names=['tax','inc','hig','dri']);
plt.savefig('regtree.png')
```

#### Feature Importances

```python
plt.figure(figsize=(16,9))
plt.title('Feature Importance')
plt.xlabel('Feature Name')
plt.ylabel('Importance %')
plt.grid()
plt.bar(x.columns, dt.feature_importances_, color='r')
```

#### Performance Analysis

```python
# MAE, MSE, MAPE?
```

```python
y_pred = dt.predict(x_test)
```

```python
result = pd.DataFrame({
    'Actual': y_test,
    'Predicted': y_pred
})
```

```python
result
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_percentage_error
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
import numpy as np
np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```
