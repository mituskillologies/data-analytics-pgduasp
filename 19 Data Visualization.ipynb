```python
import matplotlib.pyplot as plt
import numpy as np
```

#### line plot

```python
x = np.array([4,15,21,34,37,51,55,67,81,85])
y = np.array([10,16,21,34,31,56,67,78,86,99])
z = np.array([87,65,54,44,32,29,28,27,11,7])
p = np.array([65,44,34,65,76,54,33,26,55,47])
```

```python
plt.title('My first plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
plt.xlim(0,100)
plt.ylim(0,100)
plt.xticks(range(0,101,10))
plt.yticks(range(0,101,10))
plt.plot(x, y)
```

```python
plt.title('My first plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
plt.xlim(0,100)
plt.ylim(0,100)
plt.xticks(range(0,101,10))
plt.yticks(range(0,101,10))
# marker: .  o  D  >  <  v  ^  *  
# ls-   :   --  -.
plt.plot(x, y, color='maroon', marker='o', ls='--',
        lw=3, ms=10, mfc='green')
```

```python
plt.plot(x, z)
```

```python
plt.plot(x, p)
```

```python
plt.title('Comparison plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
plt.plot(x, y, color='maroon', label='2023')
plt.plot(x, z, color='orange', label='2022')
plt.legend()
```

```python
plt.title('Comparison plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
plt.ylim(0,120)
plt.plot(x, y, color='maroon', label='2023')
plt.plot(x, z, color='orange', label='2022')
plt.plot(x, p, color='green', label='2021')
plt.legend(loc='upper left')
```

```python
plt.subplot(1,2,1)
plt.plot(x, y)
plt.subplot(1,2,2)
plt.plot(x, z)
```

```python
plt.subplot(2,1,1)
plt.plot(x, y)
plt.subplot(2,1,2)
plt.plot(x, z)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(2,2,1)
plt.grid()
plt.plot(x, y, color='r')
plt.subplot(2,2,3)
plt.plot(x, z, color='g')
plt.subplot(1,2,2)
plt.title('Last Year')
plt.plot(x, p, color='magenta')
```

```python
# Dataset: ingredients.csv, movies.csv

# ingredients: x-> sweetness, y->crunchiness
# movies: x-> bugest y-> celebrities
```

```python
import pandas as pd
```

```python
ing = pd.read_csv('datasets/ingredients.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
ing['type'] = le.fit_transform(ing['type'])
```

```python
plt.scatter(ing['sweetness'], ing['crunchiness'], c=ing['type'])
```

```python
mov = pd.read_csv('datasets/movies.csv')
```

```python
mov['Class'] = le.fit_transform(mov['Class'])
```

```python
plt.scatter(mov['Budget'], 
            mov['Celebrities'], c=mov['Class'])
```

#### Bar plot

```python
stu = pd.read_csv('datasets/student.csv')
```

```python
plt.figure(figsize=(16,9))
plt.title('Student Performance')
plt.grid()
plt.xlabel('Student Name')
plt.ylabel('Marks')
plt.bar(stu['name'], stu['marks'], color='yellow')
```

```python
plt.figure(figsize=(16,9))
plt.title('Student Performance')
plt.grid()
plt.ylabel('Student Name')
plt.xlabel('Marks')
plt.barh(stu['name'], stu['marks'], color='cyan')
```

### Pie chart
